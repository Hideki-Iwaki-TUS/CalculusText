```python
# 演習2.10

import numpy as np

# 年金現価係数（数列の和の公式）を使って求めた場合
def present_value(i, n, coupon_rate, P):
    # i = 割引率，n = 満期（年），coupon_rate = クーポン・レート，P=額面金額
    coupon = coupon_rate*P
    return coupon*(1-1/(1+i)**n)/i+P/(1+i)**n

print('割引率11%のときの現在価値 =',round(present_value(0.11, 6, 0.1, 100),2),'円')
print('割引率 9%のときの現在価値 =',round(present_value(0.09, 6, 0.1, 100),2),'円')

# 数列の和の公式を使わずに求めた場合
def present_value(i, n, coupon_rate, P):
    # i = 割引率，n = 満期（年），coupon_rate = クーポン・レート，P=額面金額
    coupon = coupon_rate*P
    return coupon*np.sum(np.array([1/(1+i)**k for k in np.arange(1,n+1)]))+P/(1+i)**n

print('割引率11%のときの現在価値 =',round(present_value(0.11, 6, 0.1, 100),2),'円')
print('割引率 9%のときの現在価値 =',round(present_value(0.09, 6, 0.1, 100),2),'円')
```

```python
# 演習2.11

import numpy as np

# (2.12）式を使って求めた場合
def future_value(i, n, C, P):
    # i = 利子率，n = 満期（年），C = 積立金，P=元金
    return P*(1+i)**n + (1+i)*C*((1+i)**n-1)/i

print('10年後の元利合計金額 =',round(future_value(0.01,10,1,100),4),'万円')


# (2.10)式第一等式右辺を使って求めた場合
def future_value2(i, n, C, P):
    # i = 利子率，n = 満期（年），C = 積立金，P=元金
    a=P; r=(1+i); d=(1+i)*C
    return a*r**n+d*np.sum(np.array([r**k for k in np.arange(n)]))

print('10年後の元利合計金額 =',round(future_value2(0.01,10,1,100),4),'万円')
```

```python
# 問4.10
from sympy import *
var('n,r') # 記号の定義

# F=額面，C=クーポン
F=100
C=0.05*F

# 債券価格を金利rの関数として定義する. 
def P(r):
    return summation(C/(1+r)**n, (n, 1,5))+F/(1+r)**5

# デュレーションをrの関数として定義
D = -diff(P(r),r)*(1+r)/P(r)
D = lambdify(r, D)

# コンベクシティをrの関数として定義 
Co = diff(P(r),r,2)*(1+r)**2/P(r)
Co = lambdify(r, Co)

# (4.17)右辺をrとDelta_r(Δr)の関数として定義
def DeltaP(r,Delta_r):
    return P(r)*(-D(r)*Delta_r/(1+r)+1/2*Co(r)*(Delta_r/(1+r))**2)


print('1%上昇したときの価格変化量 =', round(DeltaP(0.05,0.01),2),'円')
print('1%下落したときの価格変化量 =', round(DeltaP(0.05,-0.01),2),'円')
```
